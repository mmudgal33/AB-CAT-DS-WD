**1. Reverse a String**  
Write a function that takes a string and returns it reversed.

**Solution:**
```javascript
function reverseString(str) {
  return str.split('').reverse().join('');
}

console.log(reverseString("hello"));  // Output: "olleh"
```

**2. Find the Longest Word in a Sentence**  
Write a function that takes a sentence and returns the longest word.

**Solution:**
```javascript
function findLongestWord(sentence) {
  const words = sentence.split(' ');
  let longestWord = '';
  for (const word of words) {
    if (word.length > longestWord.length) {
      longestWord = word;
    }
  }
  return longestWord;
}

console.log(findLongestWord("I am learning JavaScript programming"));  // Output: "programming"
```

**3. Palindrome Check**  
Write a function that checks if a string is a palindrome.

**Solution:**
```javascript
function isPalindrome(str) {
  const reversed = str.split('').reverse().join('');
  return str === reversed;
}

console.log(isPalindrome("madam"));  // Output: true
console.log(isPalindrome("hello"));  // Output: false
```

**4. Fibonacci Series**  
Write a function to generate Fibonacci numbers up to `n`.

**Solution:**
```javascript
function generateFibonacciSeries(n) {
  const series = [0, 1];
  for (let i = 2; i < n; i++) {
    series[i] = series[i - 1] + series[i - 2];
  }
  return series;
}

console.log(generateFibonacciSeries(6));  // Output: [0, 1, 1, 2, 3, 5]
```

**5. Find Duplicates in an Array**  
Write a function to find duplicates in an array of numbers.

**Solution:**
```javascript
function findDuplicates(arr) {
  const duplicates = [];
  const numCount = {};
  for (const num of arr) {
    if (numCount[num]) {
      duplicates.push(num);
    } else {
      numCount[num] = 1;
    }
  }
  return duplicates;
}

console.log(findDuplicates([1, 2, 3, 4, 5, 6, 3, 4]));  // Output: [3, 4]
```

**6. Array Sum**  
Write a function to sum all numbers in an array.

**Solution:**
```javascript
function arraySum(arr) {
  let sum = 0;
  for (const num of arr) {
    sum += num;
  }
  return sum;
}

console.log(arraySum([1, 2, 3, 4]));  // Output: 10
```

**7. Two Sum Problem**  
Write a function that returns indices of the two numbers in an array that add up to a target sum.

**Solution:**
```javascript
function twoSum(arr, targetSum) {
  const numMap = new Map();
  for (let i = 0; i < arr.length; i++) {
    const complement = targetSum - arr[i];
    if (numMap.has(complement)) {
      return [numMap.get(complement), i];
    }
    numMap.set(arr[i], i);
  }
  return [];
}

console.log(twoSum([2, 7, 11, 15], 9));  // Output: [0, 1]
```

**8. Count Vowels and Consonants in a String**  
Write a function that counts vowels and consonants in a string.

**Solution:**
```javascript
function countVowelsConsonants(str) {
  let vowels = 0, consonants = 0;
  for (const char of str.toLowerCase()) {
    if ('aeiou'.includes(char)) vowels++;
    else if (char >= 'a' && char <= 'z') consonants++;
  }
  return { vowels, consonants };
}

console.log(countVowelsConsonants("hello"));  // Output: { vowels: 2, consonants: 3 }
```

**9. Flatten an Array of Arrays**  
Write a function to flatten an array of arrays.

**Solution:**
```javascript
function flattenArray(arr) {
  return arr.flat();
}

console.log(flattenArray([1, [2, 3], [4, 5]]));  // Output: [1, 2, 3, 4, 5]
```

**10. Factorial of a Number Using Recursion**  
Write a function that calculates the factorial of a number using recursion.

**Solution:**
```javascript
function factorial(n) {
  if (n === 0) return 1;
  return n * factorial(n - 1);
}

console.log(factorial(5));  // Output: 120
```
