1. **Write a function that returns the first even number from an array.**

   ```javascript
   function findFirstEven(numbers) {
       for (let i = 0; i < numbers.length; i++) {
           if (numbers[i] % 2 === 0) {
               return numbers[i];
           }
       }
       return null;
   }
   console.log(findFirstEven([1, 3, 7, 8, 5])); // 8
   console.log(findFirstEven([1, 3, 7, 5])); // null
   console.log(findFirstEven([2, 4, 6])); // 2
   ```

2. **Create a function `getDayName` that returns the day of the week from a number (1-7).**

   ```javascript
   function getDayName(dayNumber) {
       try {
           switch (dayNumber) {
               case 1: return "Monday";
               case 2: return "Tuesday";
               case 3: return "Wednesday";
               case 4: return "Thursday";
               case 5: return "Friday";
               case 6: return "Saturday";
               case 7: return "Sunday";
               default: throw new Error("Invalid day number");
           }
       } catch (error) {
           return error.message;
       }
   }
   console.log(getDayName(1)); // "Monday"
   console.log(getDayName(7)); // "Sunday"
   console.log(getDayName(0)); // "Invalid day number"
   ```

3. **Write a function that counts how many times a character appears in a string.**

   ```javascript
   function countChar(str, char) {
       let count = 0;
       for (let i = 0; i < str.length; i++) {
           if (str[i] === char) {
               count++;
           }
       }
       return count;
   }
   console.log(countChar("javascript", "a")); // 2
   ```

4. **Write a function to reverse a string without using built-in methods.**

   ```javascript
   function reverseString(str) {
       let reversed = "";
       for (let i = str.length - 1; i >= 0; i--) {
           reversed += str[i];
       }
       return reversed;
   }
   console.log(reverseString("hello")); // "olleh"
   ```

5. **Write a function that returns a new array with the elements of the original array doubled.**

   ```javascript
   function doubleArray(arr) {
   ```

6. **Create a function `createCounter` that returns an object with methods to increment a counter and get its value.**

   ```javascript
   function createCounter() {
       let counter = 0;
       return {
           increment: function() {
               counter += 1;
           },
           getValue: function() {

   const counter = createCounter();
   counter.increment();
   console.log(counter.getValue()); // 1
   ```

7. **Write a function `mergeArrays` that merges two sorted arrays into a single sorted array.**

   ```javascript
   function mergeArrays(arr1, arr2) {
       let result = [];
       let i = 0, j = 0;
       while (i < arr1.length && j < arr2.length) {
           if (arr1[i] < arr2[j]) {
               result.push(arr1[i]);
               i++;
           } else {
               result.push(arr2[j]);
               j++;
           }
       }
       return result.concat(arr1.slice(i), arr2.slice(j));
   }
   console.log(mergeArrays([1, 3, 5], [2, 4, 6])); // [1, 2, 3, 4, 5, 6]
   ```

8. **Write a function that converts a string to title case (capitalizing the first letter of each word).**

   ```javascript
   function toTitleCase(str) {
       return str.split(" ")
                 .map(word => word.charAt(0).toUpperCase() + word.slice(1))
                 .join(" ");
   }
   console.log(toTitleCase("hello world")); // "Hello World"
   ```

9. **Write a function `sumNumbers` that returns the sum of numbers in an array.**

   ```javascript
   function sumNumbers(arr) {
       return arr.reduce((acc, num) => acc + num, 0);
   }
   console.log(sumNumbers([1, 2, 3, 4])); // 10
   ```

10. **Write a function `filterEven` that filters out even numbers from an array.**

   ```javascript
       return arr.filter(num => num % 2 === 0);
   }
   console.log(filterEven([1, 2, 3, 4, 5])); // [2, 4]
   ```

2. **What is a RESTful API, and how does it differ from SOAP?**
   - REST is stateless and operates over HTTP, while SOAP uses XML and requires more overhead, including a higher degree of security protocols. REST APIs are more flexible, whereas SOAP is known for its strict standards.

3. **What are closures in JavaScript?**
   - A closure allows a function to access variables from its lexical scope even after the function has executed. Example:
     ```js
     function outer() {
         let count = 0;
         return function inner() {
             count++;
             console.log(count);
         };
     }
     const increment = outer();
     increment(); // 1
     increment(); // 2
     ```

4. **What are the differences between `==` and `===` in JavaScript?**
   - `==` compares values with type coercion, while `===` compares both value and type without coercion.

5. **What is the purpose of middleware in Express.js?**
   - Middleware in Express.js is used to handle requests and responses, perform tasks like authentication, logging, and error handling. It’s essentially a chain of functions that modify or process requests before they reach the final route handler.
