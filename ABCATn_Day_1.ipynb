### 1. Function to Reverse a String

**Question:** Write a function that takes a string as input and returns the string reversed.  
**Example Input:** `"hello"`  
**Example Output:** `"olleh"`

```javascript
// Function to reverse a string
function reverseString(str) {
  // Split the string into an array of characters
  let charArray = str.split('');
  // Reverse the array of characters
  let reversedArray = charArray.reverse();
  // Join the reversed array back into a single string
  return reversedArray.join('');
}

// Example usage
console.log(reverseString("hello")); // Output: "olleh"
```

### 2. Function to Check if a Number is Prime

**Question:** Write a function that takes a number as input and checks if it is a prime number.  
**Example Input:** `7`  
**Example Output:** `true`

**Example Input:** `10`  
**Example Output:** `false`

```javascript
// Function to check if a number is prime
function isPrime(num) {
  // Handle edge cases
  if (num <= 1) return false; // Numbers less than 2 are not prime
  // Check for factors other than 1 and the number itself
  for (let i = 2; i < num; i++) {
    if (num % i === 0) {
      return false; // If divisible by any number, it's not prime
    }
  }
  return true; // If no divisors found, it's prime
}

// Example usage
console.log(isPrime(7));  // Output: true
console.log(isPrime(10)); // Output: false
```

### 3. Function to Find the Factorial of a Number

**Question:** Write a function that takes a number as input and returns its factorial.  
**Example Input:** `5`  
**Example Output:** `120`

```javascript
// Recursive function to find the factorial of a number
function factorial(n) {
  // Base case: factorial of 0 or 1 is 1
  if (n <= 1) return 1;
  // Recursive case: n * factorial of (n - 1)
  return n * factorial(n - 1);
}

// Example usage
console.log(factorial(5)); // Output: 120
```

### 4. Function to Convert a String to Uppercase Without Using Built-In Methods

**Question:** Write a function that takes a string as input and converts it to uppercase without using built-in methods like `toUpperCase()`.  
**Example Input:** `"hello World!"`  
**Example Output:** `"HELLO WORLD!"`

```javascript
// Function to convert a string to uppercase without using toUpperCase()
function toUpperCaseCustom(str) {
  let result = '';
  for (let char of str) {
    // Check if character is lowercase using ASCII values
    if (char >= 'a' && char <= 'z') {
      // Convert to uppercase by shifting ASCII code
      result += String.fromCharCode(char.charCodeAt(0) - 32);
    } else {
      // Append non-lowercase characters as they are
      result += char;
    }
  }
  return result;
}

// Example usage
console.log(toUpperCaseCustom("hello World!")); // Output: "HELLO WORLD!"
```

### 5. Function to Find the Largest Number in an Array

**Question:** Write a function that takes an array of numbers as input and returns the largest number in the array.  
**Example Input:** `[10, 20, 5, 30, 25]`  
**Example Output:** `30`

```javascript
// Function to find the largest number in an array
function maxNumber(arr) {
  // Initialize the max variable with the first element of the array
  let max = arr[0];
  // Loop through each number in the array
  for (let num of arr) {
    // Update max if the current number is greater
    if (num > max) {
      max = num;
    }
  }
  return max;
}

// Example usage
console.log(maxNumber([10, 20, 5, 30, 25])); // Output: 30
```
