### 1. Reverse a String Word by Word
**Problem:** Given a string, reverse each word in the sentence.

```javascript
function reverseEachWord(string) {
  return string.split(" ").reverse().join(" ");
}

const sentence = "Welcome to JavaScript Guide!";
console.log(reverseEachWord(sentence)); // Output: "Guide! JavaScript to Welcome"
```

### 4. Implement a Simple Shopping Cart
**Problem:** Create a simple shopping cart that allows adding/removing items and calculating total price.

```javascript
class Cart {
  constructor() {
    this.items = [];
  }

  addItem(item) {
    this.items.push(item);
  }

  removeItem(itemName) {
    this.items = this.items.filter(item => item.name !== itemName);
  }

  calculateTotal() {
    return this.items.reduce((total, item) => total + item.price * item.quantity, 0);
  }
}

const cart = new Cart();
cart.addItem({ name: "Laptop", price: 1000, quantity: 1 });
cart.addItem({ name: "Phone", price: 500, quantity: 2 });
console.log(cart.calculateTotal()); // Output: 2000
```

### 5. Check if Two Strings are Anagrams
**Problem:** Given two strings, return `true` if they are anagrams of each other.

```javascript
function areAnagrams(str1, str2) {
  const sorted1 = str1.split("").sort().join("");
  const sorted2 = str2.split("").sort().join("");
  return sorted1 === sorted2;
}

console.log(areAnagrams("Mary", "Army")); // true
```

### 7. Throttling Function
**Problem:** Implement a throttling function that limits how often a function can be called.

```javascript
function throttle(fn, delay) {
  let lastCall = 0;
  return function() {
    const now = Date.now();
    if (now - lastCall >= delay) {
      lastCall = now;
      fn.apply(this, arguments);
    }
  };
}

const log = () => console.log('Throttled!');
const throttledLog = throttle(log, 2000);
throttledLog(); // Will execute immediately
throttledLog(); // Will be ignored
```

### 8. Palindrome Check
**Problem:** Check if a string is a palindrome.

```javascript
function isPalindrome(str) {
  const cleanedStr = str.toLowerCase().replace(/[^a-zA-Z0-9]/g, '');
  return cleanedStr === cleanedStr.split('').reverse().join('');
}

console.log(isPalindrome("racecar")); // true
console.log(isPalindrome("hello")); // false
```

### 10. URL Shortener Service
**Problem:** Implement a simple URL shortener service using JavaScript.

```javascript
class URLShortener {
  constructor() {
    this.urls = {};
    this.baseUrl = "http://short.ly/";
    this.id = 0;
  }

  shorten(longUrl) {
    const shortUrl = this.baseUrl + this.id++;
    this.urls[shortUrl] = longUrl;
    return shortUrl;
  }

  expand(shortUrl) {
    return this.urls[shortUrl];
  }
}

const urlService = new URLShortener();
const shortUrl = urlService.shorten("https://www.example.com");
console.log(shortUrl); // Output: http://short.ly/0
console.log(urlService.expand(shortUrl)); // Output: https://www.example.com
```
