### 1. **Write a function to implement debouncing in JavaScript.**

**Solution:**
```javascript
function debounce(func, delay) {
    let timeoutId;
    return function (...args) {
        if (timeoutId) clearTimeout(timeoutId);
        timeoutId = setTimeout(() => func(...args), delay);
    };
}

// Example usage:
const logMessage = debounce(() => console.log("Debounced!"), 2000);
logMessage(); // Will only log after 2 seconds of inactivity
```

**Explanation:** Debouncing ensures that a function (e.g., an API call or event handler) is not executed too frequently by waiting until a certain delay period has passed without another call.

### 2. **Write a function to flatten a deeply nested array.**

**Solution:**
```javascript
function flattenArray(arr) {
    return arr.reduce((acc, item) => {
        return Array.isArray(item) ? acc.concat(flattenArray(item)) : acc.concat(item);
    }, []);
}

// Example usage:
const nestedArray = [1, [2, [3, [4, 5]]], 6];
console.log(flattenArray(nestedArray)); // [1, 2, 3, 4, 5, 6]
```

**Explanation:** This function recursively flattens any deeply nested array into a single-level array using `reduce` and `concat`.

### 3. **Write a function to check if a string is a palindrome, considering case-insensitivity and ignoring non-alphanumeric characters.**

**Solution:**
```javascript
function isPalindrome(str) {
    const cleanedStr = str.replace(/[^a-z0-9]/gi, '').toLowerCase();
    return cleanedStr === cleanedStr.split('').reverse().join('');
}

// Example usage:
console.log(isPalindrome("A man, a plan, a canal, Panama")); // true
```

**Explanation:** The function first removes non-alphanumeric characters using a regular expression, converts the string to lowercase, and then checks if it reads the same forwards and backwards.

### 4. **Write a function that accepts an array of promises and returns a single promise that resolves when all of them have resolved.**

**Solution:**
```javascript
function promiseAll(promises) {
    return new Promise((resolve, reject) => {
        let results = [];
        let count = 0;
        if (promises.length === 0) resolve(results);

        promises.forEach((promise, index) => {
            promise
                .then(value => {
                    results[index] = value;
                    count++;
                    if (count === promises.length) resolve(results);
                })
                .catch(reject);
        });
    });
}

// Example usage:
const promise1 = Promise.resolve(1);
const promise2 = Promise.resolve(2);
promiseAll([promise1, promise2]).then(console.log); // [1, 2]
```

**Explanation:** This custom implementation of `Promise.all` resolves when all promises in the array are resolved, returning their results in the order they were passed.

### 5. **Write a function to find the longest substring without repeating characters.**

**Solution:**
```javascript
function longestSubstring(str) {
    let map = new Map();
    let start = 0, maxLength = 0;

    for (let end = 0; end < str.length; end++) {
        if (map.has(str[end])) {
            start = Math.max(map.get(str[end]) + 1, start);
        }
        map.set(str[end], end);
        maxLength = Math.max(maxLength, end - start + 1);
    }
    return maxLength;
}

// Example usage:
console.log(longestSubstring("abcabcbb")); // 3 ("abc")
```

**Explanation:** This function uses the sliding window technique to keep track of characters and their positions, ensuring that only non-repeating characters are considered in the longest substring.

### 1. **Deep Merge Two Objects**
Write a function that deeply merges two objects, combining nested properties. If a property exists in both objects, the function should recursively merge them.

```javascript
function deepMerge(obj1, obj2) {
    const result = { ...obj1 };
    for (const key in obj2) {
        if (obj2.hasOwnProperty(key)) {
            if (obj1[key] && typeof obj1[key] === 'object' && typeof obj2[key] === 'object') {
                result[key] = deepMerge(obj1[key], obj2[key]);
            } else {
                result[key] = obj2[key];
            }
        }
    }
    return result;
}

const obj1 = { a: 1, b: { x: 2, y: 3 }, c: 4 };
const obj2 = { b: { y: 4, z: 5 }, d: 6 };
console.log(deepMerge(obj1, obj2));
```
**Example Output:**
```javascript
{ a: 1, b: { x: 2, y: 4, z: 5 }, c: 4, d: 6 }
```

### 2. **Implement a Debounce Function**
Create a debounce function that limits how often a callback can be invoked. The function should only execute after a specified delay when the user stops calling it.

```javascript
function debounce(func, wait) {
    let timeout;
    return function (...args) {
        clearTimeout(timeout);
        timeout = setTimeout(() => func(...args), wait);
    };
}

const log = debounce(() => console.log('Hello!'), 2000);
log(); // Will log 'Hello!' after 2 seconds if no other call happens within 2 seconds
```

### 3. **Find All Unique Pairs that Sum to a Target**
Write a function that finds all unique pairs of integers in an array that sum up to a target value.

```javascript
function findPairs(arr, target) {
    const pairs = new Set();
    const result = [];

    for (const num of arr) {
        const complement = target - num;
        if (pairs.has(complement)) {
            result.push([complement, num]);
        } else {
            pairs.add(num);
        }
    }

    return result;
}

console.log(findPairs([2, 7, 11, 15, -2, 3], 9));
```
**Example Output:**
```javascript
[ [ 2, 7 ], [ -2, 11 ] ]
```

### 4. **Implement Async Waterfall**
Create an async waterfall function that executes tasks in sequence, each task depending on the result of the previous one.

```javascript
function asyncWaterfall(tasks) {
    return tasks.reduce((prevTask, currentTask) => {
        return prevTask.then(currentTask);
    }, Promise.resolve());
}

const task1 = () => new Promise(resolve => setTimeout(() => { console.log('Task 1'); resolve(); }, 1000));
const task2 = () => new Promise(resolve => setTimeout(() => { console.log('Task 2'); resolve(); }, 1000));

asyncWaterfall([task1, task2]).then(() => console.log('All tasks completed!'));
```
**Example Output:**
```javascript
Task 1
Task 2
All tasks completed!
```
