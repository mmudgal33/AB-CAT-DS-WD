### 1. **Question: Write a function to flatten a nested array of arbitrary depth.**

```javascript
function flatten(arr) {
    return arr.reduce((acc, val) => Array.isArray(val) ? acc.concat(flatten(val)) : acc.concat(val), []);
}

// Example usage:
console.log(flatten([1, [2, [3, [4]]]])); // Output: [1, 2, 3, 4]
```

**Explanation:** This function uses recursion to flatten arrays that are nested to any level of depth.

### 2. **Question: Write a function to find the first non-repeating character in a string.**

```javascript
function firstUniqChar(str) {
    const count = {};
    for (let char of str) {
        count[char] = (count[char] || 0) + 1;
    }
    for (let char of str) {
        if (count[char] === 1) return char;
    }
    return null;
}

// Example usage:
console.log(firstUniqChar("swiss")); // Output: "w"
console.log(firstUniqChar("aabbcc")); // Output: null
```

**Explanation:** This function counts the occurrences of each character and returns the first character that occurs only once.

### 3. **Question: Implement a debounce function to limit how often a function is called.**

```javascript
function debounce(func, wait) {
    let timeout;
    return function(...args) {
        clearTimeout(timeout);
        timeout = setTimeout(() => func(...args), wait);
    };
}

// Example usage:
const log = () => console.log("Debounced!");
const debouncedLog = debounce(log, 2000);
debouncedLog(); // "Debounced!" will be logged only after 2 seconds if not called again in between.
```

**Explanation:** This `debounce` function ensures that the provided function is only called after a specified time delay, which is useful for limiting function calls like those in scroll or input events.

### 4. **Question: Write a function to merge two sorted arrays into one sorted array.**

```javascript
function mergeSortedArrays(arr1, arr2) {
    let i = 0, j = 0;
    let merged = [];
    while (i < arr1.length && j < arr2.length) {
        if (arr1[i] < arr2[j]) merged.push(arr1[i++]);
        else merged.push(arr2[j++]);
    }
    return merged.concat(arr1.slice(i)).concat(arr2.slice(j));
}

// Example usage:
console.log(mergeSortedArrays([1, 3, 5], [2, 4, 6])); // Output: [1, 2, 3, 4, 5, 6]
```

**Explanation:** This function merges two sorted arrays efficiently using a two-pointer technique.

### 5. **Question: Implement a function that deep clones an object, including nested objects.**

```javascript
function deepClone(obj) {
    if (obj === null || typeof obj !== 'object') return obj;
    let copy = Array.isArray(obj) ? [] : {};
    for (let key in obj) {
        if (obj.hasOwnProperty(key)) {
            copy[key] = deepClone(obj[key]);
        }
    }
    return copy;
}

// Example usage:
const original = { a: 1, b: { c: 2 } };
const clone = deepClone(original);
clone.b.c = 3;
console.log(original.b.c); // Output: 2 (original is not mutated)
```

**Explanation:** This function performs a deep clone by recursively copying properties, ensuring that nested objects are also cloned.

### 6. **Question: Write a function to check if a string is a palindrome, ignoring spaces and case sensitivity.**

```javascript
function isPalindrome(str) {
    str = str.replace(/\s+/g, '').toLowerCase();
    return str === str.split('').reverse().join('');
}

// Example usage:
console.log(isPalindrome("A man a plan a canal Panama")); // Output: true
console.log(isPalindrome("Hello World")); // Output: false
```

**Explanation:** This function removes spaces and converts the string to lowercase before checking if it is the same forwards and backwards.

### 7. **Question: Implement a function that returns the nth Fibonacci number using recursion.**

```javascript
function fibonacci(n) {
    if (n <= 1) return n;
    return fibonacci(n - 1) + fibonacci(n - 2);
}

// Example usage:
console.log(fibonacci(6)); // Output: 8
```

**Explanation:** This recursive function computes the Fibonacci sequence, where each number is the sum of the two preceding ones.

### 8. **Question: Write a function to flatten an object with nested objects into a single level object with dot notation for keys.**

```javascript
function flattenObject(obj, parent = '', res = {}) {
    for (let key in obj) {
        let prop = parent ? parent + '.' + key : key;
        if (typeof obj[key] === 'object' && obj[key] !== null) {
            flattenObject(obj[key], prop, res);
        } else {
            res[prop] = obj[key];
        }
    }
    return res;
}

// Example usage:
const obj = { a: { b: { c: 1 } }, d: 2 };
console.log(flattenObject(obj)); // Output: { 'a.b.c': 1, d: 2 }
```

**Explanation:** This recursive function flattens a nested object into a single-level object with keys representing the paths to the values.

### 9. **Question: Write a function that implements a throttle mechanism to limit the number of function calls over time.**

```javascript
function throttle(func, limit) {
    let inThrottle;
    return function(...args) {
        if (!inThrottle) {
            func(...args);
            inThrottle = true;
            setTimeout(() => inThrottle = false, limit);
        }
    };
}

// Example usage:
const log = () => console.log("Throttled!");
const throttledLog = throttle(log, 2000);
throttledLog(); // "Throttled!" will be logged only once within 2 seconds.
```

**Explanation:** This function ensures that the provided function is called at most once within the specified time limit.

### 10. **Question: Implement a function to find the longest substring without repeating characters.**

```javascript
function longestUniqueSubstring(str) {
    let map = new Map();
    let start = 0, maxLength = 0;
    for (let end = 0; end < str.length; end++) {
        if (map.has(str[end])) {
            start = Math.max(start, map.get(str[end]) + 1);
        }
        map.set(str[end], end);
        maxLength = Math.max(maxLength, end - start + 1);
    }
    return maxLength;
}

// Example usage:
console.log(longestUniqueSubstring("abcabcbb")); // Output: 3 ("abc")
console.log(longestUniqueSubstring("bbbbb")); // Output: 1 ("b")
```

**Explanation:** This function uses a sliding window technique to find the longest substring without repeating characters.

### Technical Questions:
1. **What is the difference between SQL and NoSQL databases? When would you choose one over the other?**
   - *SQL databases* are structured and use tables for storing data with defined schemas (e.g., MySQL, PostgreSQL), while *NoSQL databases* (e.g., MongoDB, Cassandra) allow unstructured data with flexible schemas.
   - Choose *SQL* when you need ACID compliance and structured relationships (e.g., financial applications). Choose *NoSQL* for scalability, large amounts of unstructured data, or when you need high availability and speed, such as for social media platforms or IoT.

2. **Can you explain what a microservice architecture is and how you would design a system using this approach?**
   - Microservices involve breaking down a monolithic application into smaller, self-contained services that handle specific business functions. Each service can be deployed and scaled independently. In your design, you'd use RESTful APIs or messaging systems for communication between services, databases for each service, and ensure data consistency using techniques like event sourcing or CQRS.

3. **What are some common strategies to scale an application for high availability?**
   - Use techniques such as horizontal scaling (adding more servers), database replication, load balancing, and caching strategies (like Redis or Memcached). You could also implement failover mechanisms and use cloud-native services like Kubernetes for auto-scaling.

4. **Explain the CAP theorem and its implications for database design.**
   - The *CAP theorem* (Consistency, Availability, Partition tolerance) states that in a distributed system, you can only guarantee two out of three properties. For example, in a system designed for high availability, you might sacrifice consistency, leading to temporary data inconsistencies during network partitions.

5. **What are the trade-offs involved in choosing between synchronous and asynchronous processing in distributed systems?**
   - *Synchronous* processing ensures immediate consistency but can result in performance bottlenecks due to waiting for responses. *Asynchronous* processing improves throughput and system responsiveness, but it can introduce complexity in managing eventual consistency and handling failures.
