### 1. **Find the Factorial of a Number**

**Problem:** Write a function that calculates the factorial of a number.

```javascript
function factorial(n) {
    if (n === 0) return 1;
    return n * factorial(n - 1);
}

// Example 1:
console.log(factorial(5)); // Output: 120 (5*4*3*2*1)

// Example 2:
console.log(factorial(0)); // Output: 1
```

**Solution Explanation:**  
- The factorial of a number `n` is the product of all integers from `n` down to `1`.
- This is implemented recursively.

### 2. **Check if a Number is Prime**

**Problem:** Write a function that checks if a number is prime.

```javascript
function isPrime(num) {
    if (num <= 1) return false;
    for (let i = 2; i <= Math.sqrt(num); i++) {
        if (num % i === 0) return false;
    }
    return true;
}

// Example 1:
console.log(isPrime(7)); // Output: true

// Example 2:
console.log(isPrime(10)); // Output: false
```

**Solution Explanation:**  
- A prime number is a number greater than `1` that is divisible only by `1` and itself.
- The loop checks for factors up to the square root of the number to optimize performance.

### 3. **Merge Two Sorted Arrays**

**Problem:** Write a function to merge two sorted arrays into one sorted array.

```javascript
function mergeSortedArrays(arr1, arr2) {
    let result = [];
    let i = 0, j = 0;
    
    while (i < arr1.length && j < arr2.length) {
        if (arr1[i] < arr2[j]) result.push(arr1[i++]);
        else result.push(arr2[j++]);
    }
    
    return result.concat(arr1.slice(i), arr2.slice(j));
}

// Example 1:
console.log(mergeSortedArrays([1, 3, 5], [2, 4, 6])); // Output: [1, 2, 3, 4, 5, 6]

// Example 2:
console.log(mergeSortedArrays([1, 10, 20], [5, 15])); // Output: [1, 5, 10, 15, 20]
```

**Solution Explanation:**  
- The function uses two pointers, `i` and `j`, to traverse both arrays and merge them in sorted order.

### 4. **Remove Duplicates from an Array**

**Problem:** Write a function that removes duplicate values from an array.

```javascript
function removeDuplicates(arr) {
    return [...new Set(arr)];
}

// Example 1:
console.log(removeDuplicates([1, 2, 2, 3, 4, 4])); // Output: [1, 2, 3, 4]

// Example 2:
console.log(removeDuplicates([5, 6, 7, 7, 8])); // Output: [5, 6, 7, 8]
```

**Solution Explanation:**  
- `new Set(arr)` creates a set from the array, automatically removing duplicates.
- The spread operator `...` converts the set back into an array.

### 5. **Sum of All Even Numbers in an Array**

**Problem:** Write a function that returns the sum of all even numbers in an array.

```javascript
function sumEvenNumbers(arr) {
    return arr.filter(num => num % 2 === 0).reduce((sum, num) => sum + num, 0);
}

// Example 1:
console.log(sumEvenNumbers([1, 2, 3, 4, 5])); // Output: 6 (2 + 4)

// Example 2:
console.log(sumEvenNumbers([10, 20, 30, 40])); // Output: 100
```

**Solution Explanation:**  
- Use `filter()` to select even numbers and `reduce()` to sum them.

### 6. **Flatten an Array of Arbitrary Depth**

**Problem:** Write a function that flattens an array of arbitrary depth (nested arrays of varying depths).

```javascript
function flattenDeep(arr) {
    return arr.reduce((acc, val) => acc.concat(Array.isArray(val) ? flattenDeep(val) : val), []);
}

// Example 1:
console.log(flattenDeep([1, [2, [3, 4]], 5])); // Output: [1, 2, 3, 4, 5]

// Example 2:
console.log(flattenDeep([10, [20, 30], 40])); // Output: [10, 20, 30, 40]
```

**Solution Explanation:**  
- Recursively flatten the array using `reduce()` and `concat()`.

### 7. **Find the First Non-Repeating Character**

**Problem:** Write a function that finds the first non-repeating character in a string.

```javascript
function firstNonRepeatingChar(str) {
    for (let i = 0; i < str.length; i++) {
        if (str.indexOf(str[i]) === str.lastIndexOf(str[i])) {
            return str[i];
        }
    }
    return null;
}

// Example 1:
console.log(firstNonRepeatingChar("swiss")); // Output: "w"

// Example 2:
console.log(firstNonRepeatingChar("aabbcc")); // Output: null
```

**Solution Explanation:**  
- `indexOf()` and `lastIndexOf()` are used to find if a character appears more than once.

### 8. **Find the Longest Word in a String**

**Problem:** Write a function that finds the longest word in a string.

```javascript
function findLongestWord(str) {
    let words = str.split(' ');
    return words.reduce((longest, word) => word.length > longest.length ? word : longest, '');
}

// Example 1:
console.log(findLongestWord("The quick brown fox jumped over the lazy dog")); // Output: "jumped"

// Example 2:
console.log(findLongestWord("hello world")); // Output: "hello"
```

**Solution Explanation:**  
- Split the string into words and use `reduce()` to find the longest word.

### 9. **Count Vowels in a String**

**Problem:** Write a function that counts the number of vowels in a string.

```javascript
function countVowels(str) {
    return str.match(/[aeiouAEIOU]/g) ? str.match(/[aeiouAEIOU]/g).length : 0;
}

// Example 1:
console.log(countVowels("hello")); // Output: 2

// Example 2:
console.log(countVowels("JavaScript")); // Output: 3
```

**Solution Explanation:**  
- Use a regular expression `/[aeiouAEIOU]/g` to match all vowels.
- `match()` returns an array of matches, and its `length` gives the count.

### 10. **Find the Missing Number in a Sequence**

**Problem:** Write a function that finds the missing number in a sequence of consecutive numbers.

```javascript
function findMissing(arr) {
    const n = arr.length + 1;
    const totalSum = (n * (n + 1)) / 2;  // Sum of numbers from 1 to n
    const arrSum = arr.reduce((sum, num) => sum + num, 0);
    return totalSum - arrSum;
}

// Example 1:
console.log(findMissing([1, 2, 4, 5, 6])); // Output: 3

// Example 2:
console.log(findMissing([10, 11, 12, 14])); // Output: 13
```

**Solution Explanation:**  
- Calculate the sum of all numbers from 1 to `n` using the formula for the sum of an arithmetic series.
- Subtract the sum of the array from the total sum to find the missing number.
