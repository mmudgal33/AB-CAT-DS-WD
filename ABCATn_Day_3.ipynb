### 1. **Reverse a String**  
Write a function to reverse a string.  
**Solution:**
```javascript
function reverseString(str) {
  return str.split('').reverse().join('');
}
console.log(reverseString("hello")); // "olleh"
```

### 2. **Find the Longest Word in a Sentence**  
Write a function that returns the longest word in a sentence.  
**Solution:**
```javascript
function findLongestWord(sentence) {
  const words = sentence.split(' ');
  let longestWord = '';
  for (const word of words) {
    if (word.length > longestWord.length) {
      longestWord = word;
    }
  }
  return longestWord;
}
console.log(findLongestWord("The quick brown fox jumps over the lazy dog.")); // "jumps"
```

### 3. **Check Palindrome**  
Write a function to check if a string is a palindrome.  
**Solution:**
```javascript
function isPalindrome(str) {
  const reversed = str.split('').reverse().join('');
  return str === reversed;
}
console.log(isPalindrome("madam")); // true
```

### 4. **Generate Fibonacci Series**  
Write a function that generates a Fibonacci series up to a given number.  
**Solution:**
```javascript
function generateFibonacciSeries(n) {
  const series = [0, 1];
  for (let i = 2; i < n; i++) {
    series.push(series[i - 1] + series[i - 2]);
  }
  return series;
}
console.log(generateFibonacciSeries(5)); // [0, 1, 1, 2, 3]
```

### 5. **Sum of Array Elements**  
Write a function to sum all numbers in an array.  
**Solution:**
```javascript
function arraySum(arr) {
  return arr.reduce((sum, num) => sum + num, 0);
}
console.log(arraySum([1, 2, 3, 4])); // 10
```

### 6. **Find Duplicates in an Array**  
Write a function that returns any duplicates in an array.  
**Solution:**
```javascript
function findDuplicates(arr) {
  const numCount = {};
  const duplicates = [];
  arr.forEach(num => {
    if (numCount[num]) {
      duplicates.push(num);
    } else {
      numCount[num] = 1;
    }
  });
  return duplicates;
}
console.log(findDuplicates([1, 2, 3, 1, 4, 2])); // [1, 2]
```

### 7. **Two Sum**  
Given an array and a target sum, find two numbers in the array whose sum equals the target sum.  
**Solution:**
```javascript
function twoSum(arr, target) {
  const map = new Map();
  for (let i = 0; i < arr.length; i++) {
    const complement = target - arr[i];
    if (map.has(complement)) {
      return [map.get(complement), i];
    }
    map.set(arr[i], i);
  }
  return [];
}
console.log(twoSum([2, 7, 11, 15], 9)); // [0, 1]
```

### 8. **Remove Duplicates from Array**  
Write a function that removes duplicates from an array.  
**Solution:**
```javascript
function removeDuplicates(arr) {
  return [...new Set(arr)];
}
console.log(removeDuplicates([1, 2, 2, 3, 4, 4])); // [1, 2, 3, 4]
```

### 9. **Find the Missing Number in Array**  
Given an array of consecutive numbers, find the missing number.  
**Solution:**
```javascript
function findMissingNumber(arr) {
  const n = arr.length + 1;
  const expectedSum = (n * (n + 1)) / 2;
  const actualSum = arr.reduce((sum, num) => sum + num, 0);
  return expectedSum - actualSum;
}
console.log(findMissingNumber([1, 2, 3, 5])); // 4
```

### 10. **Deep Clone an Object**  
Write a function that deep clones an object.  
**Solution:**
```javascript
function deepClone(obj) {
  return JSON.parse(JSON.stringify(obj));
}
const original = { a: 1, b: { c: 2 } };
const cloned = deepClone(original);
cloned.b.c = 3;
console.log(original.b.c); // 2
console.log(cloned.b.c); // 3
```

### Technical Questions:
1. **How would you approach designing a fault-tolerant system?**
   - A fault-tolerant system ensures the service continues to operate correctly even when one or more components fail. It involves identifying potential failure points, using redundancy for critical components, and implementing load balancing, data replication, and automated failover mechanisms to ensure high availability.

2. **What are the differences between functional and object-oriented programming?**
   - In **functional programming**, functions are first-class citizens and emphasize immutability and statelessness. **Object-oriented programming** focuses on objects that encapsulate state and behavior. Key concepts include inheritance and polymorphism.

3. **How would you scale a system like Twitter to handle millions of users?**
   - Scaling a system like Twitter would involve strategies such as maintaining a **cache** for users' feeds, **load balancing** for even traffic distribution, and implementing **asynchronous queues** for scalable message consumption.

4. **How do you ensure high availability in distributed systems?**
   - High availability is achieved through **redundancy**, **load balancing**, and **replication** to prevent single points of failure. **Automated failover mechanisms** and continuous monitoring are also key to detect and respond to issues promptly.
